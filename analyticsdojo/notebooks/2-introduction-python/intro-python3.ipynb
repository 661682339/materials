###Python is Interpreted
Python is an interpreted language, in contrast to Java and C which are compiled languages.
This means we can type statements into the interpreter and they are executed immediately. This works well for data analytics because we can build things step by step.
Simply press play at the end of a cell to execute it.

```python
5 + 5
```

### Variables can be Assigned Directly
- Set the variable and conduct basic operations.  It is not requred to declare the variable type prior to assigning the value.

- If an output isn't assigned, it will be provided as output.

- The output can be specified with the `print()` command.

```python
x = 5
y = 'plus 25 is'
z = x + 25
```

```python
z
```

```python
print('Variables:', x, y, z)
```

### Assignments versus equations

- In Python when we write `x = 5` this means something different from an equation $x=5$.

- Unlike variables in mathematical models, variables in Python can refer to different things as more statements are interpreted.


```python
a = 1
print ('The value of x is ', a)

a = 2.5
print ('Now the value of x is ', a)

a = 'hello there'
print ('Now it is ', a)
```
